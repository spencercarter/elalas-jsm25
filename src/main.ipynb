````python
import json
import pandas as pd
garage = pd.read_json('./data/GaRAGe_benchmark.jsonl',
                          lines=True,           # Treats each line as a separate JSON object
    encoding='utf-8',     # Specify encoding
    chunksize=None        # Read entire file at once (default)
)

garage.head(10)
````

````python
sample_record = garage.sample(n=1).iloc[0]

# Print with custom formatting
print("Sampled Record:")
print("=" * 60)

for field_name, field_value in sample_record.items():
    print(f"Field: {field_name}")
    print(f"Value: {field_value}")
    print("-" * 40)
````
